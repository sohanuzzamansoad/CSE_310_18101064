# 2.Write the code of all the classes:

### Food Class:

```
class Food:
  def __init__(self,name, food_type):
    self.name=name
    self.food_type=food_type

  def show_details(self):
    print("I am "+self.name)
```

### Chinese Food Class:

```
class Chinese_Food(Food):

  def __init__(self,name,food_type):
    self.cuisine_name='Chinese'
    Food.__init__(self,name,food_type)

  def show_details(self):
    print("I am "+self.name+" From "+self.cuisine_name+" Food")
```

### Japanese Food Class:

```
class Japanese_Food(Food):

  def __init__(self,name,food_type):
    self.cuisine_name='Japanese'
    Food.__init__(self,name,food_type)

  def show_details(self):
    print("I am "+self.name+" From "+self.cuisine_name+" Food")
```

### American Food Class:

```
class American_Food(Food):

  def __init__(self,name,food_type):
    self.cuisine_name='American'
    Food.__init__(self,name,food_type)

  def show_details(self):
    print("I am "+self.name+" From "+self.cuisine_name+" Food")
```

### American-Japanese Food Class:

```
class AmericanJapanese_Food(American_Food, Japanese_Food ):

  def __init__(self,name,food_type):
    self.cuisine_name='American-Japanese'
    Food.__init__(self,name,food_type)


#Method Overriding
  def show_details(self):
    print("I am "+self.name+" From "+self.cuisine_name+" Food")
```

### Sichuan Pork Class:

```
class Sichuan_Pork(Chinese_Food):
  def __init__(self,name, boiled_level=1 ):
    self.boiled_level=boiled_level
    Chinese_Food.__init__(self,name,"Main Dish")
  
  def Madeof(self):
    print("Made of: Sichuan Spice and Pork")

  def BoiledLevel(self):
    print("Boiled Level: "+str(self.boiled_level))
```

### Sushi Class:

```
class Sushi(Japanese_Food):
  def __init__(self,name, ingred,level_rice=1):
    self.ingred=ingred
    self.level_rice=level_rice
    Japanese_Food.__init__(self,name,"Appetizer, Main Dish")
  
  def Madeof(self):
    print("Made of: Vinegared Rice with "+self.ingred)

  def RiceLevel(self):
    if self.level_rice==1:
      print("Ratio of Rice is Normal")
    else:
      print("Ratio of Rice is High")
  
```

###  Cobb Salad Class:

```
class CobbSalad(American_Food):
  def __init__(self,name, ingred):
    self.ingred=ingred
    American_Food.__init__(self,name,"Salad")
  
  def Madeof(self):
    print("Made of: chopped salad greens, tomato, crisp bacon with "+self.ingred)

  
```

### Ramen-Burger Class:

```
class Ramen_Burger(AmericanJapanese_Food):
  def __init__(self,name, spicy_level=1):
    self.spicy_level=spicy_level
    AmericanJapanese_Food.__init__(self,name,"Snacks")
  
  def Madeof(self,spicy_level):
    if self.spicy_level==1:
      print("Made of: Japanese Spicy Ramen & American Burger")
    else:
      print("Made of: Japanese Super Spicy Ramen & American Burger")  

#Method Overloading
  
  def Madeof(self):
    print("Made of: Japanese Ramen & American Burger")


```
